```python
def mongo_to_pandas(collections):
    """Convert MongoDB collection to Pandas DataFrame.
    
    Parameters
    ----------
    collections : MongoDB collection
    
    Returns
    -------
    Pandas DataFrame Object representing the collection
    """
    return pd.DataFrame.from_records(collections.find())
```

```python
# get rid of documents without ratings and review_counts
def drop_rows_na(df):
    """Drop rows with missing values in name, ratings, review_count
    
    Parameters
    ----------
    df : Pandas DataFrame
    
    Returns
    -------
    Pandas DataFrame 
        Dataframe without missing critical values
    
    """
    return df.dropna(subset=['name', 'ratings', 'review_count'])
```

```python
def drop_rows_duplicates(df):
    """Drops duplicated rows
    
    Due to MongoDB's default _id for each insert, we have to check duplicates on
    columns: name, rating, review_count
    
    Parameters
    ----------
    df: Pandas DataFrame
    
    Returns
    -------
    Pandas DataFrame
        Dataframe without any duplicates
    """
    return df.drop_duplicates(subset=['name','ratings','review_count'])
```

```python
def explode_column(df, column, threshold=0):
    """Creates new columns from dictionary keys and values from dictionary values.
    
    The function coverts columns (pd.Series) into a list type and gets
    passed to the pd.DataFrame() function to create a new dataframe. This is
    then reattached to the original dataframe passed. If the key does not exist,
    a None value will be placed instead.
    
    Parameters
    ----------
    df: Pandas DataFrame
    column: String
        Name of column with the dictionary
    threshold: Float (between 0 and 1)
        Minimum number of records that must contain the key
    
    Returns
    -------
    Pandas DataFrame
        Dataframe with new columns added
    
    """
    keys = [] # gather all the keys (various dictionaries)
    df[column].map(lambda x: keys.extend(list(x.keys())) if x != None
                   else _ )
    
    filtered_keys = [key for key in set(keys) if keys.count(key) > threshold ] # filter keys with threshold
    final_keys = list(set(filtered_keys))
    
    for i in range(len(final_keys)): # for each key make column assigned with value
        df[final_keys[i]] = df[column].map(lambda x: x.get(final_keys[i]) if x != None
                                     else None)
    
    
    
```
